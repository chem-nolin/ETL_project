```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Importing the csv file from desktop
csv_file = "flipkart.csv" #remove this if use the #comment below
flipkart = pd.read_csv(csv_file) #pd.read_csv("flipkart.csv")
```

```python
# Ascertaining the length of the file and showing the columns/data format
print(len(flipkart))
flipkart.head(15)
#flipkart.describe() #always look at the statistics
```

```python
# Remove missing values that are demarked by NaN, this won't remove blanks (update: there weren't any NaN data)
flipkart.dropna(inplace = True) 
print(len(flipkart))
flipkart.head(10)
```

```python
# This splits the first and last names apart into two separate columns and in the process creates a new dataframe (table) separate from the orginal.
# Normally n here would be n=1, but is n=4 here as there blank spaces around the names.
new_flipkart = flipkart["flipkart_author"].str.split("\s", n = 4, expand = True)
new_flipkart.head(20)
```

```python
# This command is adding the new first and last name columns in the above dataframe to the original dataframe. 
# The [#] refers to the position in the above dataframe
flipkart["last_name"]= new_flipkart[3] 
flipkart["first_name"]= new_flipkart[2] 

flipkart.head(30)
```

```python
# Correcting the title column, so they all have the same "title" format...
# Here 'str.title' could also have been str.lower, str.upper, str.capitalize, or str.swapcase
new2_flipkart = flipkart['flipkart_title'].str.title()
new2_flipkart.head()
```

```python
# Adding this new "Title" column with the correct title text format to the original dataframe
flipkart["title"]= new2_flipkart 
flipkart.head()
```

```python
# Converting the price from rupees to dollars, current exchange rate is 1 Rupee = 0.015 USD
flipkart['flipkart_price'] = flipkart['flipkart_price'].apply(lambda x: int(x*0.015))
flipkart.head()
```

```python
# Dropping orginal (old) name and title columns from the dataframe
flipkart.drop(columns =["flipkart_author", "flipkart_title"], inplace = True)
flipkart.head()
```

```python
# Renaming the remaining original column names
renamed_flipkart = flipkart.rename(columns={"flipkart_isbn10":"isbn10", "flipkart_ratings count":"flipkart_ratings_count", "flipkart_price":"flipkart_price_usd", "flipkart_stars":"flipkart_rating"})
renamed_flipkart.head()
```

```python
# Reordering all the columns
reorganized_flipkart = renamed_flipkart[["last_name","first_name", "isbn10", "title", "flipkart_price_usd", "flipkart_rating", "flipkart_ratings_count"]]
reorganized_flipkart.head()
```

```python
reorganized_flipkart.to_csv("reorganized_flipkart.csv", index=False, header=True)
```

```python
# Re-Importing the super clean csv file from desktop. 
csv_file_2 = "flipkart_for_reimport.csv"
flipkart_2 = pd.read_csv(csv_file_2, encoding="ISO-8859-1") # have to use this encoding="ISO-8859-1" to prevent errors
```
